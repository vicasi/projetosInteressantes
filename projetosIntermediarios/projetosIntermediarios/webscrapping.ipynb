```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
from datetime import datetime


def procurar_emprego():
    agora = datetime.now().strftime('%H:%M')
    print(agora)
    perguntar_habilidades = input('liste suas habilidades: ')
    print('digite o intervalo de tempo em que o loop de buscas deve atuar(digite 0 caso deseje anular determinado momento)')
    h = int(input('horas: '))
    m = int(input('minutos: '))
    s = int(input('segundos: '))
    total = s + m * 60 + h * 3600
    driver = webdriver.Firefox()
    while agora < '23:59':
        if agora == '00:00':
            lista_vagas_dia = open('lista_de_vagas_ultimo_dia', 'w')
            lista_vagas_dia.write(f'{companhia}\n{habilidades_companhia}\n{data_postagem}')
            return
        html = 'https://www.timesjobs.com/'
        print('entrou no loop')
        driver.get(html)
        driver.implicitly_wait(30)
        pesquisa = driver.find_element_by_id('txtKeywords')
        pesquisa.send_keys(perguntar_habilidades, Keys.RETURN)
        wait = WebDriverWait(driver, 30)
        carregamento_cookie = wait.until(
            EC.presence_of_element_located((By.CLASS_NAME, "blkclor"))
        )
        cookies = driver.find_element(By.XPATH, '//button[@class="waves-effect waves-light btn gdpr-btn"]').click()
        carregamento_completo = wait.until(
            EC.presence_of_all_elements_located((By.XPATH, '//li[@class="clearfix job-bx wht-shd-bx"]'))
        )
        page_source = driver.page_source
        soup = BeautifulSoup(page_source, 'lxml')
        li = soup.find_all('li', class_='clearfix job-bx wht-shd-bx')
        duas_horas = open('lista_de_vagas_ultimas_2_horas', 'a')
        for vaga in li:
            companhia = vaga.find('h3', class_='joblist-comp-name').text
            data_postagem = vaga.find('span', class_='sim-posted').text
            if 'few' in data_postagem:
                habilidades = vaga.find('span', class_='srp-skills').text.replace(' ', '').replace(',', '/')
                habilidades_companhia = vaga.find('span', class_='srp-skills').text.replace(" ", "").replace(",","/")
                print(f'\n-- companhia: {companhia}\n-- habilidades: {habilidades}')
                duas_horas.write(f'{companhia}\n{habilidades_companhia}\n{data_postagem}')  
    time.sleep(total)
if __name__ == '__main__':
    deseja = input('deseja procurar novo emprego?(s/n): ')
    while deseja == 's':
        procurar_emprego()
    
```
